```python
import numpy as mp
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = mp.sin(x)

plt.plot(x, y)
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-1, 1)
plt.xlabel('x')
plt.ylabel('y = sin(x)')
plt.title('Función seno')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 10*np.pi, 1000)

y_sin = np.sin(x)
y_cos = np.cos(x)


plt.scatter(x, y_sin, s=10, c=x, cmap='hsv')
plt.plot(x, y_cos, c='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funciones seno y coseno')
plt.show()
```

```python
from matplotlib_venn import venn2
import matplotlib.pyplot as plt


set1 = set([1, 2, 3, 4, 5])
set2 = set([4, 5, 6, 7, 8])

venn2([set1, set2], ('Conjunto 1', 'Conjunto 2'))

plt.title('Diagrama de Venn')
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x  = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contour(x, y, z)
plt.show()
```
